# Приклад 1

## Package

```r
# install.packages("tidyverse")                           # Install & load tidyverse
library(tidyverse)
```

## Load example data

```r
data(PlantGrowth) 
```

```r
?PlantGrowth
```

## Density plot using base R

```r
plot(density(PlantGrowth$weight))
```

## Density plot using ggplot2

```r
ggplot(data = PlantGrowth,                                # Density plot using ggplot2
       aes(x = weight)) +
  geom_density()
```

## Grouped density plot

```r
ggplot(data = PlantGrowth,                                # Grouped density plot
       aes(x = weight,
           color = group)) +
  geom_density()
```

## Overlapping densities

```r
ggplot(data = PlantGrowth,                                # Overlapping densities
       aes(x = weight,
           fill = group)) +
  geom_density(alpha = 0.3)
```

## Group data using dplyr

```r
PlantGrowth_grouped <- PlantGrowth %>%                    # Group data using dplyr
  group_by(group) %>% 
  summarize(mean_weight = mean(weight))
PlantGrowth_grouped
```

## ggplot2 bar chart

```r
ggplot(data = PlantGrowth_grouped,                        # ggplot2 bar chart
       aes(x = group,
           y = mean_weight)) +
  geom_col()
```

## Manipulation & visualization

```r
PlantGrowth %>%
  group_by(group) %>% 
  summarize(mean_weight = mean(weight)) %>% 
  ggplot(aes(x = group,
             y = mean_weight)) +
  geom_col()
```
