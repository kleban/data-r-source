# Естетичне відображення (Aesthetic Mapping)

У цьому модулі розглядається естетичне відображення — ключове поняття в ggplot2, яке пов’язує змінні даних із візуальними властивостями, такими як колір, розмір і форма. Ви дізнаєтесь, як ефективно використовувати естетичні відображення для створення інформативних і візуально привабливих графіків, які підкреслюють закономірності та взаємозв’язки у ваших даних.

```r
# install.packages("ggplot2")                             # Install & load ggplot2
library(ggplot2)
```

```r
?ToothGrowth
```

```r
data(ToothGrowth)                                         # Load example data
ToothGrowth
```

```r
ggplot(data = ToothGrowth,                                # Basic aesthetic mapping
       aes(y = len)) +
  geom_boxplot()
```

```r
ggplot(data = ToothGrowth,                                # Mapping supp to x-axis
       aes(x = supp,
           y = len)) +
  geom_boxplot()
```

```r

ggplot(data = ToothGrowth,                                # Mapping supp to fill
       aes(x = supp, 
           y = len, 
           fill = supp)) +
  geom_boxplot()
```

```r
set.seed(54698)                                           # Seed for reproducibility
 
ggplot(data = ToothGrowth,                                # Mapping supp to color
       aes(x = supp, 
           y = len, 
           color = supp)) +
  geom_jitter(width = 0.2, 
              size = 4)
```

```r
ggplot(data = ToothGrowth,      # Mapping dose to size
       aes(x = supp, 
           y = len, 
           size = dose)) +
  geom_jitter(width = 0.2)
```

```r

 

 

 
ggplot(data = ToothGrowth,                                # Combined aesthetic mapping
       aes(x = supp, 
           y = len, 
           color = supp, 
           shape = supp, 
           size = dose)) +
  geom_jitter(width = 0.2)
 
ggplot(data = ToothGrowth,                                # Mapping inside geom
       aes(x = supp, 
           y = len)) +
  geom_jitter(aes(color = supp, 
                  shape = supp, 
                  size = dose), 
              width = 0.2)
```
