# The Grammar of Graphics

У цьому модулі розглядається граматика графіки (Grammar of Graphics) — основа ggplot2. Ви дізнаєтесь, як ця структурована система розкладає візуалізації на ключові компоненти, що спрощує створення та налаштування графіків у систематичний спосіб. Розуміння цих принципів підвищить вашу здатність створювати ефективні та змістовні візуалізації.

## Package

```r
# install.packages("ggplot2")                             # Install & load ggplot2
library(ggplot2)
```

## Data

```r
data(PlantGrowth)
PlantGrowth
```

```r
ggplot(data = PlantGrowth)        # data
```

```r
ggplot(data = PlantGrowth,       # Aesthetics
       aes(y = weight))
```

```r
 ggplot(data = PlantGrowth,       # Geometries
       aes(y = weight)) +
  geom_boxplot()
```

```r
ggplot(data = PlantGrowth,  # Facets
       aes(y = weight)) +
  geom_boxplot() +
  facet_wrap(~group)
```

```r
ggplot(data = PlantGrowth,      # Statistics
       aes(y = weight)) +
  geom_boxplot() +
  facet_wrap(~ group) +
  stat_summary(aes(x = 0),
               fun = mean,
               geom = "point",
               color = "red",
               size = 5)
```

```r
ggplot(data = PlantGrowth,                                # Coordinates
       aes(y = weight)) +
  geom_boxplot() +
  facet_wrap(~ group) +
  stat_summary(aes(x = 0),
               fun = mean,
               geom = "point",
               color = "red",
               size = 3) +
  coord_flip()
```

```r
ggplot(data = PlantGrowth,                                # Theme
       aes(y = weight)) +
  geom_boxplot() +
  facet_wrap(~ group) +
  stat_summary(aes(x = 0),
               fun = mean,
               geom = "point",
               color = "red",
               size = 3) +
  coord_flip() +
  theme_bw()
```
